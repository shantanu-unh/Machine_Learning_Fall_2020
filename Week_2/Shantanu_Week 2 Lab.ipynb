# 1. Create four DataFrames from the four .csv files

```python
import pandas as pd

df_q1 = pd.read_csv('Divvy_Trips_2018_Q1.csv')
df_q2 = pd.read_csv('Divvy_Trips_2018_Q2.csv')
df_q3 = pd.read_csv('Divvy_Trips_2018_Q3.csv')
df_q4 = pd.read_csv('Divvy_Trips_2018_Q4.csv')
```

# 2. Create a single dataframe by combining the four previously created

hint: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html


```python
df_combined = pd.concat([df_q1, df_q2, df_q3, df_q4])
df_combined.head(3)
```

# 5. Turn the two date columns into type pandas.datetime 

```python
df_combined['start_time'] = pd.to_datetime(df_combined['start_time'])
df_combined['end_time'] = pd.to_datetime(df_combined['end_time'])

df_combined
```

# 7. Create a new column with the month of start_time

```python
# df_combined.head(3)

df_combined['new_column'] = df_combined['start_time'].dt.month

df_combined.tail()

# df_combined['new_column']
```
